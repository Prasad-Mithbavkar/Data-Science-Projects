# Naive bayes

    GaussianNB: For datasets with continuous features.
    MultinomialNB: For datasets with discrete features (e.g., text data with word counts).
    BernoulliNB: For datasets with binary features (e.g., presence or absence of a feature).
    ComplementNB: Similar to MultinomialNB, but handles imbalanced classes better.

# Naive bayes with hyperparameters

```python

# Create different Naive Bayes classifiers
gaussian_model = GaussianNB()
multinomial_model = MultinomialNB()
bernoulli_model = BernoulliNB()
complement_model = ComplementNB()

# Define hyperparameters and their potential values for each model
param_grid_gaussian = {
    # No hyperparameters to tune for GaussianNB
}

param_grid_multinomial = {
    'alpha': [0.01, 0.1, 1.0],
    'fit_prior': [True, False]
}

param_grid_bernoulli = {
    'alpha': [0.01, 0.1, 1.0],
    'binarize': [0.0, 0.5, 1.0],
    'fit_prior': [True, False]
}

param_grid_complement = {
    'alpha': [0.01, 0.1, 1.0],
    'fit_prior': [True, False],
    'norm': [True, False]
}

# GridSearchCV with cross-validation for each model
grid_search_gaussian = GridSearchCV(gaussian_model, param_grid_gaussian, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_multinomial = GridSearchCV(multinomial_model, param_grid_multinomial, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_bernoulli = GridSearchCV(bernoulli_model, param_grid_bernoulli, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_complement = GridSearchCV(complement_model, param_grid_complement, cv=5, scoring='accuracy', n_jobs=-1)

# Fit and evaluate models
grid_search_gaussian.fit(X_train, y_train)
grid_search_multinomial.fit(X_train, y_train)
grid_search_bernoulli.fit(X_train, y_train)
grid_search_complement.fit(X_train, y_train)

best_params_gaussian = grid_search_gaussian.best_params_
best_params_multinomial = grid_search_multinomial.best_params_
best_params_bernoulli = grid_search_bernoulli.best_params_
best_params_complement = grid_search_complement.best_params_

best_gaussian_model = GaussianNB(**best_params_gaussian)
best_multinomial_model = MultinomialNB(**best_params_multinomial)
best_bernoulli_model = BernoulliNB(**best_params_bernoulli)
best_complement_model = ComplementNB(**best_params_complement)

best_gaussian_model.fit(X_train, y_train)
best_multinomial_model.fit(X_train, y_train)
best_bernoulli_model.fit(X_train, y_train)
best_complement_model.fit(X_train, y_train)

y_pred_gaussian = best_gaussian_model.predict(X_test)
y_pred_multinomial = best_multinomial_model.predict(X_test)
y_pred_bernoulli = best_bernoulli_model.predict(X_test)
y_pred_complement = best_complement_model.predict(X_test)

accuracy_gaussian = accuracy_score(y_test, y_pred_gaussian)
accuracy_multinomial = accuracy_score(y_test, y_pred_multinomial)
accuracy_bernoulli = accuracy_score(y_test, y_pred_bernoulli)
accuracy_complement = accuracy_score(y_test, y_pred_complement)

print(f"Gaussian Naive Bayes Best Hyperparameters:", best_params_gaussian)
print(f"Gaussian Naive Bayes Accuracy: {accuracy_gaussian:.4f}")

print(f"Multinomial Naive Bayes Best Hyperparameters:", best_params_multinomial)
print(f"Multinomial Naive Bayes Accuracy: {accuracy_multinomial:.4f}")

print(f"Bernoulli Naive Bayes Best Hyperparameters:", best_params_bernoulli)
print(f"Bernoulli Naive Bayes Accuracy: {accuracy_bernoulli:.4f}")

print(f"Complement Naive Bayes Best Hyperparameters:", best_params_complement)
print(f"Complement Naive Bayes Accuracy: {accuracy_complement:.4f}")
```
